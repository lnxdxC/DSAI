# L06.1 Logistic regression 
In the following we will explain the logistic regression by using the sklearn implementation.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X = pd.read_csv('car_data.csv')
```

```python
fig, ax = plt.subplots(figsize=(10, 8))
ax.scatter(X.Age, X.Class, c='#79AEA3', marker='x', s=100)
ax.set_yticks([0,1])
ax.set_yticklabels(['False', 'True'])
ax.set_xlabel('Age in years')
ax.set_ylim([-0.2,1.2])
fig.savefig('logit_reg_1.png', bbox_inches='tight', transparent=True)
```

```python
# Fit a model to it
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()

# Training/Fitting the Model
regressor.fit(X.Age.values.reshape(len(X.Age.values), 1), X.Class.values)

regressor.predict(X.Age.values.reshape(len(X.Age.values), 1))
```

```python
fig, ax = plt.subplots(figsize=(10, 8))
ax.scatter(X.Age, X.Class, c='#79AEA3', marker='x', s=100)
ax.plot(X.Age.values.reshape(len(X.Age.values), 1), regressor.predict(X.Age.values.reshape(len(X.Age.values), 1)), alpha=0.75, color='#DB5461')
ax.set_yticks([0,1])
ax.set_yticklabels(['False', 'True'])
ax.set_xlabel('Age in years')
ax.set_ylim([-0.2,1.2])
fig.savefig('logit_reg_2.png', bbox_inches='tight', transparent=True)
```

```python
fig, ax = plt.subplots(figsize=(10, 8))
ax.scatter(X.Age, X.Class, c='#79AEA3', marker='x', s=100)
ax.plot(X.Age.values.reshape(len(X.Age.values), 1), regressor.predict(X.Age.values.reshape(len(X.Age.values), 1)), alpha=0.75, color='#DB5461')
ax.plot([X.Age.min(), X.Age.max()], [0.5,0.5], linestyle='dashed', color=[0.5,0.5,0.5], linewidth=0.5) 

ax.plot([42.4, 42.4], [0, 1], linestyle='dashed', color=[0.5,0.5,0.5], linewidth=0.5) 


ax.set_yticks([0,1])
ax.set_yticklabels(['False', 'True'])
ax.set_xlabel('Age in years')
ax.set_ylim([-0.2,1.2])
fig.savefig('logit_reg_3.png', bbox_inches='tight', transparent=True)
```

```python
def logit(x):
    return 1 / (1 + np.exp(-x))
```

```python
x = np.linspace(-5,5,100)
y = logit(x)

fig, ax = plt.subplots(2, figsize=(10,8), sharex=True)
ax[0].plot(x,x, color='#79AEA3', label='linear function')
ax[0].set_xlim([-5,5])
ax[0].legend()

ax[1].plot(x,y, color='#79AEA3', label='logit function')
ax[1].set_xlim([-5,5])
ax[1].legend()

fig.savefig('logit_function.png', bbox_inches='tight', transparent=True)
```
