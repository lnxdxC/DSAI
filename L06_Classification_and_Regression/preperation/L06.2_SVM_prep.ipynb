```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x1 = np.random.uniform(0, 11, 100)
y1 = np.random.uniform(0, 5, 100)

x2 = np.random.uniform(0, 11, 100)
y2 = np.random.uniform(6, 11, 100)
```

```python
fig, ax = plt.subplots(figsize=(10,10))

ax.scatter(x1, y1, s=100, color='#79AEA3', label='Class I')
ax.scatter(x2, y2, s=100, color='#DB5461', label='Class II', marker='x')

#ax.plot([0,11], [5, 5], color='#C1C95E', label='Support Vector I')
#ax.plot([0,11], [6, 6], color='#43DAD3', label='Support Vector II')

#ax.plot([0,11], [5.5, 5.5], color='#6B5ABF', linewidth=3, label='Hyperplain')

ax.set_ylabel('Feature II')
ax.set_xlabel('Feature I')

ax.legend()

fig.savefig('SVM_0.png', bbox_inches='tight', transparent=True)
```

```python
X = [[-1,1,-1,1], [-1,1,1,-1]]
y = [-1,-1,1,1]
```

```python
fig, ax = plt.subplots(figsize=(10,10))

ax.scatter(X[0], X[1], c=y, s=100)

ax.set_ylabel('Feature II')
ax.set_xlabel('Feature I')

fig.savefig('XOR.png', bbox_inches='tight', transparent=True)
```

```python
def transform(X):
    return np.sqrt(np.square(np.add(X[0], X[1])))
```

```python
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(projection='3d')

ax.scatter3D(X[0], X[1], transform(X), s=100,c=y, alpha=1)

ax.set_xlabel('Feature I')
ax.set_ylabel('Feature II')
ax.view_init(azim=-90, elev=90)
ax.set_zticks([])

fig.savefig('XOR_1.png', bbox_inches='tight', transparent=True)
```

```python
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(projection='3d')

ax.scatter3D(X[0], X[1], transform(X), s=100, c=transform(X), alpha=1)

ax.set_xlabel('Feature I')
ax.set_ylabel('Feature II')
ax.set_zlabel('Transform')
ax.view_init(azim=22, elev=22)

x = np.array([[1, 1], [-1, -1]])
y = np.array([[1, -1], [1, -1]])
z = np.array([[1, 1], [1, 1]])

ax.plot_surface(x, y, z, alpha=0.5)

fig.savefig('XOR_2.png', bbox_inches='tight', transparent=True)
```
