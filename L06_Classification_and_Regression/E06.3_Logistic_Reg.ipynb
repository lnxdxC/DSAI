# L06.3 Logistic regression 
In the following we will explain the logistic regression by using the sklearn implementation.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

## Data import

```python
X = pd.read_csv('https://raw.githubusercontent.com/lnxdxC/DSAI/main/L06_Classification_and_Regression/car_data.csv')
```

```python
fig, ax = plt.subplots()
ax.scatter(X.Age, X.Class, c='#79AEA3', marker='x', s=100)
ax.set_xlabel('Age in years')
```

### Prepare the data by means of the *train_test_split* method

```python
X_train, X_test, y_train, y_test = train_test_split(X[['Age']], X.Class, test_size=0.1)
```

### Create the Logistic Regression

In the following, we:
1. Create an instance of the `LogisticRegeression`
2. Learn our model by the training data set by means of the `.fit()`-method
3. Apply the model on our test data set by means of the `.predict()`-method
4. Evaluate the performance of the classification (accuracy) by means of the `.score()`-method

```python
model = LogisticRegression()  # Object on the LogisticRegression class
```

```python
model.fit(X_train, y_train)  # As usual in sklearn: .fit trains the model by means of the parsed data
```

```python
model.predict(X_test)  # Apply the model in our test data set. The result is the classification of the X_test data
```

```python
model.score(X_test, y_test)  # Evaluate the performance of the model by parsing the ground truth information to the prediction
```

#### Let`s draw the resulting logit function on our data

```python
fig, ax = plt.subplots(figsize=(10,8))

ax.scatter(X.Age, X.Class, c='#79AEA3', marker='x', s=100, label='data')
ax.set_xlabel('Age in years')
ax.set_ylabel('Probability in %')

Xs = np.linspace(X.Age.min(), X.Age.max(), 20)
Ys = [model.predict_proba([[value]])[0][1] for value in Xs]

ax.plot(Xs, Ys, color='#DB5461', label='logistic model')
ax.legend()
```
